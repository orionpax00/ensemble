```python
import os
import sys

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
BASE_FILE = "../data/base/base.csv"
MAIN_FOLDER = "../data/main/"
```

```python
##Load base file
df = pd.read_csv(BASE_FILE)
df.head()
```

```python
df.describe()
```

```python
##Droping all row that doen't have numerical values
df = df.dropna()
df.describe()
```

```python
df_ = df.convert_objects(convert_numeric=True)
plotdf = df_[["DT", "Solar Radiation", "Sunshine hour", "MAX Temp", "EVP","Vapour press","wind speed","PM2.5","PM10","NO2"]]
plotdf.plot(subplots=True,figsize=(22,12))
plt.show()
```

```python
sns.pairplot(plotdf, diag_kind='hist', hue="wind speed" )
```

```python
## drop all empty valued rows and saving to main file

df_[['PM2.5','PM10','NO2']].replace('', np.nan)
df_.dropna(subset=['PM2.5','PM10','NO2'], inplace=True)
df_.to_csv(os.path.join(MAIN_FOLDER,"main.csv"))
```

```python
plotdf = df_[["DT", "Solar Radiation", "Sunshine hour", "MAX Temp", "EVP","Vapour press","wind speed","PM2.5","PM10","NO2"]]
plotdf.plot(subplots=True,figsize=(24,12))
plt.show()
```
