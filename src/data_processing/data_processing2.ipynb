```python
#import some important modules
import os
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colors
import seaborn as sns
import numpy as np
```

```python
#add the file saved in last data processing file
df = pd.read_csv('../../filtered_data_file/2000_main_data.csv')
new_df_not_imp = df.loc[:,['Month','Hour','Day']]
new_df_imp = df.drop(columns=['Month','Hour','Day'])
```

As this data is for pre hour solar radiation fall on the earth so we can draw per weak, per month, per day, per year heatmap like plots

```python
# Per day for every month
ghi = new_df_imp.pop('GHI')
```

```python
# per day for jan
data_jan = []
for i in range(31):
    data_jan.append(ghi[i*24:(i+1)*24])
    
data_feb = []
for i in range(31,60):
    data_feb.append(ghi[i*24:(i+1)*24])
    
data_mar = []
for i in range(60,91):
    data_mar.append(ghi[i*24:(i+1)*24])
    
data_apr = []
for i in range(91,121):
    data_apr.append(ghi[i*24:(i+1)*24])
    
data_may = []
for i in range(121,152):
    data_may.append(ghi[i*24:(i+1)*24])
    
data_june = []
for i in range(152,182):
    data_june.append(ghi[i*24:(i+1)*24])
    
data_july = []
for i in range(182,213):
    data_july.append(ghi[i*24:(i+1)*24])
    
data_aug = []
for i in range(213,244):
    data_aug.append(ghi[i*24:(i+1)*24])
    
data_sep = []
for i in range(244,274):
    data_sep.append(ghi[i*24:(i+1)*24])
    
data_oct = []
for i in range(274,305):
    data_oct.append(ghi[i*24:(i+1)*24])
    
data_nov = []
for i in range(305,335):
    data_nov.append(ghi[i*24:(i+1)*24])
    
data_dec = []
for i in range(335,365):
    data_dec.append(ghi[i*24:(i+1)*24])
# data_jan = np.asarray(data_jan)
```

```python
fig, ax = plt.subplots(ncols=3,nrows=4,figsize=(18,24))
ax[0,0].pcolor(data_jan, cmap='seismic', vmin=min(ghi),vmax=max(ghi))
ax[0,1].pcolor(data_feb, cmap='seismic', vmin=min(ghi),vmax=max(ghi))
ax[0,2].pcolor(data_mar, cmap='seismic', vmin=min(ghi),vmax=max(ghi))
ax[1,0].pcolor(data_apr, cmap='seismic', vmin=min(ghi),vmax=max(ghi))
ax[1,1].pcolor(data_may, cmap='seismic', vmin=min(ghi),vmax=max(ghi))
ax[1,2].pcolor(data_june, cmap='seismic', vmin=min(ghi),vmax=max(ghi))
ax[2,0].pcolor(data_july, cmap='seismic', vmin=min(ghi),vmax=max(ghi))
ax[2,1].pcolor(data_aug, cmap='seismic', vmin=min(ghi),vmax=max(ghi))
ax[2,2].pcolor(data_sep, cmap='seismic', vmin=min(ghi),vmax=max(ghi))
ax[3,0].pcolor(data_oct, cmap='seismic', vmin=min(ghi),vmax=max(ghi))
ax[3,1].pcolor(data_nov, cmap='seismic', vmin=min(ghi),vmax=max(ghi))
ax[3,2].pcolor(data_dec, cmap='seismic', vmin=min(ghi),vmax=max(ghi))
plt.show()
```

```python
## for hour and year
data_h_year = []
for i in range(365):
    data_h_year.append(ghi[i*24:(i+1)*24])
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.pcolor(np.asarray(data_h_year), cmap='seismic', vmin=min(ghi),vmax=max(ghi))
plt.show()
```
